# Cleaning up the data 

My dataset: https://www.kaggle.com/datasets/starblasters8/color-distribution-of-sour-patch-kids?resource=download


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df= pd.read_csv("SourPatchData.csv")

# Did creating the dataframe work?
df.head(5)
```

```python
#Missing Data
#Check for Nulls

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Plot a heatmap of any missing data
cols = df.columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colours))
#This confirms data is not missing
```

```python
#Irregular Data
#Histogram to look for outliers
df.hist()
```

```python
#Looking at Green a littler closer

sns.displot(data=df, x='green', height = 5, aspect = 3)
```

```python
#Use the describe() method
df.describe()

#looks like Red shows up the most with a min of 2 and a max of 12
```

```python
#Boxplot
df.boxplot()

#Red and Yellow have some outliers
```
