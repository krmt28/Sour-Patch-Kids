## Sour Patch Kids
The distribution of colors in each 2 oz. Sour Patch Kids bag.
Hypothesis is that distributions of colors will be consistent.  
My dataset: https://www.kaggle.com/datasets/starblasters8/color-distribution-of-sour-patch-kids?resource=download


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df= pd.read_csv("SourPatchData.csv")

# Did creating the dataframe work?
df
```

```python
#check size of my table - rows and columns
df.shape
```

```python
#look at data types in the database
df.info()
```

```python
#check for nulls
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#get statitical summary 
df.describe()
```

Lets see if visualizing the data shows us anything

```python

#look for correlation - heatmap
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), annot=True, cbar=True, cmap="Spectral", fmt=".1f")

plt.savefig('heatmap.png')

#not great visualization for this data - no conclusions
```
